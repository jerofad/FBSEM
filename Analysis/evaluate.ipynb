```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
# File paths
train_path = '/home/fdgjer001/FBSEM/json (1).json'
val_path = '/home/fdgjer001/FBSEM/json.json'

# Load data
with open(train_path, 'r') as f:
    train_data = json.load(f)

with open(val_path, 'r') as f:
    val_data = json.load(f)
```

```python
# Extract epochs and losses
train_epochs = [item[1] for item in train_data]
train_loss = [item[2] for item in train_data]

val_epochs = [item[1] for item in val_data]
val_loss = [item[2] for item in val_data]
```

```python


# Plot
plt.figure()
plt.plot(train_epochs, train_loss, label='Train Loss')
plt.plot(val_epochs, val_loss, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training vs Validation Loss')
plt.legend()

# Save and show
plot_path = 'loss_plot.png'
plt.savefig(plot_path)
plt.show()

plot_path
```

```python
# --- Plot 1: remove epoch 0 ---
train_epochs_no0 = train_epochs[1:]
train_loss_no0 = train_loss[1:]

val_epochs_no0 = val_epochs[1:]
val_loss_no0 = val_loss[1:]

plt.figure()
plt.plot(train_epochs_no0, train_loss_no0, label='Train Loss')
plt.plot(val_epochs_no0, val_loss_no0, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss (Epoch 1 onward)')
plt.legend()

plot1_path = 'loss_plot_no_epoch0.png'
plt.savefig(plot1_path)
plt.show()
plt.close()
```

```python
# --- Plot 2: log scale ---
plt.figure()
plt.plot(train_epochs, train_loss, label='Train Loss')
plt.plot(val_epochs, val_loss, label='Validation Loss')
plt.yscale('log')
plt.xlabel('Epoch')
plt.ylabel('Loss (log scale)')
plt.title('Training vs Validation Loss (Log Scale)')
plt.legend()

plot2_path = 'loss_plot_log.png'
plt.savefig(plot2_path)
plt.show()
plt.close()
```
